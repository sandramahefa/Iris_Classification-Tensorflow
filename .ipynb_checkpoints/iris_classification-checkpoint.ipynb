```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data_brut = pd.read_csv("iris.csv")
data_brut.head()
```

```python
x = data_brut.iloc[:,0:4].values
x = np.array(x)
print("data label \n", x[0:1])
y = data_brut.iloc[:,4:].values
y = np.array(y)
print("data target \n", y[0])
```

```python
#TRansform the outpout into an array of 3 colmns because our model have à 3 outpout to predict
y_set = np.array([[1,0,0] for i in range(50)])
y_ver = np.array([[0,1,0] for i in range(50)])
y_vir = np.array([[0,0,1] for i in range(50)])
y = np.concatenate((y_set,y_ver,y_vir),axis=0)
y.dtype
```

```python
#transform y into np.ndarray
y=np.vstack(y).astype(np.float)
```

```python
y.dtype
```

```python
#Préprocessing
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
#split train data and test data
x_train ,x_test, y_train, y_test = train_test_split(x,y, test_size=0.1, random_state=42)
print("x_train \n", x_train)
print("y_train \n", y_train)
#len(x_test)
```

# CONSTRUCT THE MODEL WITH TENSORFLOW

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(units=6, activation=tf.nn.relu,kernel_initializer="uniform",input_dim=4))
```

```python
model.add(tf.keras.layers.Dense(units=3,activation="softmax",kernel_initializer='uniform'))
```

```python
model.compile(optimizer='adam',loss="categorical_crossentropy",metrics="accuracy")
```

```python
model.summary()
```

```python
#Train the model with 10.000 epochs
model.fit(x_train,y_train,epochs=1000)
```

```python
# We have an 0.9333 accuracy in the end of the training
test_loss, test_accuracy = model.evaluate(x_test,y_test)
print(f'test loss {test_loss} et test accuracy {test_accuracy}')
```
